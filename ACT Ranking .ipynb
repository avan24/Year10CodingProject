```python


import matplotlib.pyplot as plt
    
ACT = ['<24','25-27','28-30','31-33','34-36']
Bar = [200,200,200,200,200]

#shrink the width of the graph
f = plt.figure()
f.set_figwidth(2.9)
plt.subplots(figsize=(4, 5)) 


plt.barh(ACT,Bar) 
plt.title('Average ACT Score', fontweight="bold")
plt.xticks([]) #hides x-axis
plt.ylabel('ACT Score Range', fontsize='14')


Uni1 = ("Harvard, MIT, Stanford")# the text that I will be plotting
plt.text(10, 3.9, Uni1) # plot the position of the text on the coordinate (x,y)

Uni2 = ("UCLA, UVA, UF")
plt.text(10, 2.9, Uni2)

Uni3 = ("CLEM, OSU")
plt.text(10, 1.9, Uni3)
        
Uni4 = ("USF, MSU, LSU")
plt.text(10, 0.9, Uni4)

Uni5 = ("CU, VSU")
plt.text(10, -0.1, Uni5)

plt.show()



```
